```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import h5py
import os
import math
```

## Iterative Training - retrain entire flow each time we add another flow layer

```python
savedir = '/scratch/jmj/dgm_maxout40/synthetic/E1_8_b100_init3_annealed_iterative'

stage = []
logs = []
k=2
t=20
filename = 'q_k%s_t%s.h5' % (k,t)
with h5py.File(os.path.join(savedir,filename),'r') as Q:
    zz = np.array(Q['z%s' % 1]).astype('float')

a = np.histogram2d(zz[:,1],zz[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
plt.pcolor(a[1],a[2],a[0])
plt.title(filename,fontsize=20)

```

```python
savedir = '/scratch/jmj/dgm_maxout40/synthetic/E1_8_b100_init3_annealed_iterative/'

len_normflow = 3
max_epochs = 100

stage = []
KL = []
NLL = []
for k in range(1,len_normflow+1):
    
    with h5py.File(os.path.join(savedir,'q_k%s_t0.h5' % k),'r') as Q:
        z0 = []
        for i in range(k+1):
            z0.append(np.array(Q['z%s' % i]).astype('float'))
    
    with h5py.File(os.path.join(savedir,'q_k%s_t%s.h5' % (k,max_epochs)),'r') as Q:
        z100 = []
        for i in range(k+1):
            z100.append(np.array(Q['z%s' % i]).astype('float'))
    
    z = [z0,z100]
    stage.append(z)

    with h5py.File(os.path.join(savedir,'stage%s_log.h5' % k),'r') as L:
        kl = np.array(L['eval_KL']).astype('float')
        nll = np.array(L['eval_nll']).astype('float')

    KL.append(kl)
    NLL.append(nll)
```

```python
ncols = 10
nrows = 2*len_normflow
plt.figure(1,figsize=(15,int(1.5*nrows)))
for s,z in enumerate(stage):
    z0 = z[0]
    z100 = z[1]
    
    for k,z0_k in enumerate(z0):
        a = np.histogram2d(z0_k[:,1],z0_k[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
        plt.subplot(nrows,ncols,2*s*ncols+k+2)
        plt.pcolor(a[1],a[2],a[0])
    
    plt.subplot(nrows,ncols,(2*s+1)*ncols+1)
    plt.plot(KL[s],'b')
    plt.plot(NLL[s],'g--')
    
    for k,z100_k in enumerate(z100):
        a = np.histogram2d(z100_k[:,1],z100_k[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
        plt.subplot(nrows,ncols,(2*s+1)*ncols+k+2)
        plt.pcolor(a[1],a[2],a[0])
        
plt.tight_layout()
```
