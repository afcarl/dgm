### visualize histogram

```python
def flow(u,w,b,z0):
    wTu = np.dot(u,w)
    m = -1 + np.log(1+np.exp(wTu))
    u_hat = u + (m-wTu)*w/(np.dot(w,w)+1e-12)
    h = np.tanh(z0*w.reshape(1,-1)+b)
    uh = u_hat.reshape(1,-1)*h
    
    psi = (1-h**2)*w
    logdetJ = -np.log(np.abs(1+np.sum(u.reshape(1,-1)*psi,axis=1)))
    return z0 + uh, logdetJ

def E1(z):
    val1 = 0.5*((la.norm(z,axis=1)-2)/0.4)**2
    val2 = np.exp(-0.5*((z[:,0]-2)/0.6)**2)
    val3 = np.exp(-0.5*((z[:,0]+2)/0.6)**2)
    val4 = -np.log(val2+val3)
    return val1 + val4
```

```python
z0 = np.random.randn(100000,2)
z, kl = flow(u=np.array([0,1.1]),w=np.array([0,15]),b=0,z0=z0)
nll = E1(z)
F = nll+kl
```

```python
a = np.histogram2d(z[:,1],z[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
plt.pcolor(a[1],a[2],a[0])
```

## Visualize objective function over b, u, w

```python
uu = np.arange(-15,15.01,0.25)
ww = np.arange(-20.0,20.01,0.25)
bb = np.arange(-0.2,0.21,0.1)

FF = []
NLL = []
z0 = np.random.randn(100,2)
for b in bb:
    F = np.zeros((len(uu),len(ww)))
    nll = np.zeros((len(uu),len(ww)))
    for i,u in enumerate(uu):
        for j,w in enumerate(ww):
            z, kl = flow(u=np.array([u,0]),w=np.array([w,0]),b=b,z0=z0)
            nll[i,j] = np.mean(E1(z))
            F[i,j] = nll[i,j]+np.mean(kl)
    FF.append(F)
    NLL.append(nll)
```

```python
Fmin = min([np.min(F) for F in FF])
print "Fmin = %s" % Fmin
```

#### upperbound

```python
ncols = 5
nrows = np.ceil(len(bb)*1.0/ncols)
plt.figure(1,figsize=(15,int(2*nrows)))
for i,b in enumerate(bb):
    #logFF = np.log(-Fmin + 1 + FF[i])
    plt.subplot(nrows,ncols,i+1)
    plt.pcolor(uu,ww,-FF[i].T,vmin=-8,vmax=-Fmin)
    plt.title('b=%0.3f' % b)
    plt.ylim(min(ww),max(ww))
    plt.xlim(min(uu),max(uu))
    plt.ylabel('w')
    plt.xlabel('u')
plt.tight_layout()
```
