```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import h5py
import os
import math
```

## 100 random initializations

```python
len_normflow = 1

num_trials = 100

max_epochs = 60

z = []
KL = []
NLL = []
params = []
for t in range(1,num_trials+1):
    savedir = '/scratch/jmj/dgm_maxout40/synthetic/E1_1_b100_init3_annealed_trials/trial%s' % t
    
    zt = []
    
    with h5py.File(os.path.join(savedir,'q_t0.h5'),'r') as Q:
        zt.append(np.array(Q['z%s' % 1]).astype('float'))
    
    with h5py.File(os.path.join(savedir,'q_t%s.h5' % max_epochs),'r') as Q:
        zt.append(np.array(Q['z%s' % 1]).astype('float'))
    
    z.append(zt)
    
    with h5py.File(os.path.join(savedir,'log.h5'),'r') as L:
        KL.append(np.array(L['eval_KL']).astype('float'))
        NLL.append(np.array(L['eval_nll']).astype('float'))
        params.append(np.array(L['params']).astype('float'))

```

```python
ncols = 10
nrows = 2*int(math.ceil(num_trials*1.0/ncols))
plt.figure(1,figsize=(15,int(1.5*nrows)))
for i in range(num_trials):
    zt = z[i]
    zt0 = zt[0]
    zt1 = zt[1]
    
    row = int(math.floor(i*1.0/ncols))
    col = (i % 10) + 1
    
    #a = np.histogram2d(zt0[:,1],zt0[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
    plt.subplot(nrows,ncols,2*row*ncols + col)
    #plt.pcolor(a[1],a[2],a[0])
    plt.plot(KL[i],'b',label='KL')
    plt.plot(NLL[i],'g--',label='NLL')
    plt.ylim(-5,5)
    
    a = np.histogram2d(zt1[:,1],zt1[:,0],bins=[100,100],range=[[-4,4],[-4,4]])
    plt.subplot(nrows,ncols,(2*row+1)*ncols + col)
    plt.pcolor(a[1],a[2],a[0])
    #plt.title('z%s' % i,fontsize=24)
plt.tight_layout()
```

```python
kls = np.array([KL[i][-1] for i in range(num_trials)])
nlls = np.array([NLL[i][-1] for i in range(num_trials)])
bimodal = (np.logical_and(nlls > 1.5,nlls <=3.5)).sum()
unimodal = (nlls <= 1.5).sum()
nan = (nlls > 3.5).sum() + np.isnan(nlls).sum()

print 'bimodal: %s ~ %0.1f%%' % (bimodal,bimodal*100.0/num_trials)
print 'unimodal: %s ~ %0.1f%%' % (unimodal,unimodal*100.0/num_trials)
print 'nan: %s ~ %0.1f%%' % (nan,nan*100.0/num_trials)
```
