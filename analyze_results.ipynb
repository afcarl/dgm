```python
%matplotlib inline

import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def getResults(path):
    results = h5py.File(path,'r')
    train = np.array(results['train'])
    test = np.array(results['test'])
    test_logpv = np.array(results['test_logpv'])
    results.close()
    return {'train':train,'test':test,'test_logpv':test_logpv}
```

```python
def plotResults(x,label=''):
    y = np.array(x).ravel()
    plt.plot(y,label=label)
    
def plotAll(results,split):
    for k,v in sorted(results.iteritems(),key=lambda x:x[0]):
        plotResults(v[split],'%0.2f %s' % (min(v[split]),k))
```
